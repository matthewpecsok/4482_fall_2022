I would like to provide some background of adopting and comparisons of cloud-based Interative Development Environments (IDEs) for data mining and data science courses the instructor.  

Based on years of teaching IS 6482 - Introductory Data Mining for Business Analytics and IS 6493 - Learning from Data Science at Scale and IS 6483 - Advanced Data Mining, the challenges of setting up IDEs on students' own desktops or laptops could persist and hence impeded on students' learning of the main course content in those years. The challenges mainly stem from (1) students' unique computing environments' configurations, (2) the limitations of the instructor and teaching assistant (TAs) in helping resolve students' installation problems without intruding or being perceived to intrude on students' privacy, and (3) the unexpected software updates during a course. These challenges often delayed students' learning especially at the beginning of a course and caused students' and the instructor's stress and time to overcome the obstacles that arose. The biggest challenge which tended to happen once or twice in a course is when multiple installation attempts eventually failed and an alternative was not readily available for a student to complete an assignment or even to take an exam. 

After researching about the different cloud-based IDE environments, I adopted Databricks' Community Edition for teaching IS 6493 entirely in PySpark in 2018 and RStudio Cloud for teaching R in 2020. The abilities for the students to quickly jump start the IDE for learning and practicing course knowledge really made a significant difference ever since then. 

In the last year, I have been impressed by the flexibility and performance of Colab for developing and evaluating machine learning models. Colab offers 12 GB of memory on 1 GPU (or TPU) in a free account. The upload and code execution speeds have been superior to those in RStudio Cloud and Databricks. However, beyond this course, Colab is not adequate for medium-sized and large-sized deep learning model development which typically requires paid cloud accounts or multiple advanced GPUs on a computer.
